A few random experiments with distributitons

```python
import numpy as np
import seaborn as sns
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt
```

```python
a = np.random.normal(1,0.5,10000)
b = np.random.normal(1.5,0.5,10000)
df = pd.DataFrame({"A" : a, "B" : b})
df = df.melt()
df.columns = ["Dataset","Value"]
sns.kdeplot(x="Value",hue="Dataset",data=df)
```

```python
ttest_ind(df.query("Dataset == 'A'").Value,df.query("Dataset == 'B'").Value)
```

```python
a = np.random.normal(1,0.1,1000)
b = np.random.normal(1.5,0.1,1000)
df = pd.DataFrame({"A" : a, "B" : b})
df = df.melt()
df.columns = ["Dataset","Value"]
sns.kdeplot(x="Value",hue="Dataset",data=df)
```

```python
ttest_ind(df.query("Dataset == 'A'").Value,df.query("Dataset == 'B'").Value)
```

```python
figure, axes = plt.subplots(1,2,sharex=True,sharey=True)
for idx,sd in enumerate([0.1,1]):
    a = np.random.normal(1,sd,1000)
    b = np.random.normal(2,sd,1000)
    df = pd.DataFrame({"A" : a, "B" : b})
    df = df.melt()
    df.columns = ["Dataset","Value"]
    ax = sns.histplot(x="Value",hue="Dataset",data=df,ax=axes[idx])
```

```python
sd = 10
a = np.random.normal(1,sd,1000)
b = np.random.normal(2,sd,1000)
df = pd.DataFrame({"A" : a, "B" : b})
df = df.melt()
df.columns = ["Dataset","Value"]
ax = sns.histplot(x="Value",hue="Dataset",data=df)
ttest_ind(a,b)
```
